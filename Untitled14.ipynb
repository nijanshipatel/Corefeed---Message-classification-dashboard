```python
import pandas as pd
from datasets import Dataset

# Load your augmented dataset
df = pd.read_csv("/content/augmented_messages_2000.csv")

# Hugging Face Dataset format
dataset = Dataset.from_pandas(df)

# Train-test split
dataset = dataset.train_test_split(test_size=0.2)
train_ds, test_ds = dataset["train"], dataset["test"]
```

```python
from transformers import DistilBertTokenizerFast, DistilBertForSequenceClassification

tokenizer = DistilBertTokenizerFast.from_pretrained("distilbert-base-uncased")

def preprocess(batch):
    return tokenizer(batch["message"], truncation=True, padding="max_length")

train_ds = train_ds.map(preprocess, batched=True)
test_ds = test_ds.map(preprocess, batched=True)

# Labels: High, Normal, Low
num_labels = df["label"].nunique()
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=num_labels)
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"
```

```python
# Map string labels to integers
label2id = {label: i for i, label in enumerate(df["label"].unique())}
id2label = {i: label for label, i in label2id.items()}

df["priority"] = df["label"].map(label2id)   # new numeric label column
```

```python
dataset = Dataset.from_pandas(df[["message", "priority"]])

dataset = dataset.train_test_split(test_size=0.2)
train_ds, test_ds = dataset["train"], dataset["test"]

def preprocess(batch):
    return tokenizer(batch["message"], truncation=True, padding="max_length")

train_ds = train_ds.map(preprocess, batched=True)
test_ds = test_ds.map(preprocess, batched=True)

train_ds = train_ds.rename_column("priority", "labels")
test_ds = test_ds.rename_column("priority", "labels")
train_ds.set_format("torch", columns=["input_ids", "attention_mask", "labels"])
test_ds.set_format("torch", columns=["input_ids", "attention_mask", "labels"])
```

```python
# Load model
num_labels = len(label2id)
model = DistilBertForSequenceClassification.from_pretrained(
    "distilbert-base-uncased",
    num_labels=num_labels,
    id2label=id2label,
    label2id=label2id,
)
```

```python
from transformers import TrainingArguments, Trainer
import numpy as np
from sklearn.metrics import accuracy_score, f1_score

def compute_metrics(pred):
    labels = pred.label_ids
    preds = np.argmax(pred.predictions, axis=1)
    return {
        "accuracy": accuracy_score(labels, preds),
        "f1": f1_score(labels, preds, average="weighted"),
    }

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",   # ✅ compatible with older versions
    save_strategy="epoch",
    logging_dir="./logs",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    num_train_epochs=3,
    weight_decay=0.01,
    report_to=[]  # ✅ disables W&B logging
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=test_ds,
    processing_class=tokenizer,   # ✅ replaces deprecated tokenizer arg
    compute_metrics=compute_metrics,
)
```

```python
# 5. Train the model
# ---------------------------
trainer.train()
```

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

text = "Urgent: Meeting with CEO at 9AM. Please confirm."
inputs = tokenizer(text, return_tensors="pt").to(device)  # ✅ move inputs to GPU
outputs = model(**inputs)
pred = outputs.logits.argmax(-1).item()

print("Predicted priority:", id2label[pred])
```

```python
test_texts = [
    "Payment failed! Please update your bank details immediately.",
    "Emergency: Server is down, action required within 30 minutes.",
    "Doctor’s appointment confirmed for tomorrow at 8 AM.",
    "Lunch at 1 PM? Let me know.",
    "Your Amazon order will be delivered by Wednesday.",
    "Don’t forget to bring your notebook for class today.",
    "Congratulations! You’ve won a free iPhone. Click here.",
    "Earn $500 daily from home with this trick!",
    "SALE! 70% off on shoes, today only."
]

for text in test_texts:
    inputs = tokenizer(text, return_tensors="pt").to(device)
    outputs = model(**inputs)
    pred = outputs.logits.argmax(-1).item()
    print(f"Message: {text}\nPredicted Priority: {id2label[pred]}\n")
```

```python
%%writefile dashboard.py
import torch
from transformers import BertTokenizer, BertForSequenceClassification
import streamlit as st

# Load model and tokenizer
MODEL_PATH = "/content/priority_message_model"
tokenizer = BertTokenizer.from_pretrained(MODEL_PATH)
model = BertForSequenceClassification.from_pretrained(MODEL_PATH)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Function to predict priority
def get_priority(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True).to(device)
    outputs = model(**inputs)
    pred = outputs.logits.argmax(-1).item()
    return pred

# Streamlit app
st.title("📊 Priority Message Dashboard")
st.write("Paste your messages and see them ranked by importance")

# Input messages
user_input = st.text_area("Enter messages (one per line):",
"""
Urgent: Meeting with CEO at 9AM. Please confirm.
Lunch with friend at 1PM.
Reminder: Submit project by tomorrow.
Don't forget to water the plants.
Flight ticket confirmation for tonight.
""")

if st.button("Rank Messages"):
    messages = user_input.strip().split("\n")
    ranked = [(msg, get_priority(msg)) for msg in messages]
    ranked = sorted(ranked, key=lambda x: x[1], reverse=True)  # higher = more important

    st.subheader("📌 Ranked Messages (Highest Priority First)")
    for i, (msg, score) in enumerate(ranked, start=1):
        st.write(f"**{i}. {msg}** (Priority: {score})")
```
